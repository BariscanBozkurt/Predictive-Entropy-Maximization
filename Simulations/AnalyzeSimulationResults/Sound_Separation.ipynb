```python
import sys
sys.path.insert(0, '../Results')
sys.path.insert(0, '../../src')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import os
from scipy import stats

from python_utils.visualization_utils import create_summary_table, plot_snr_performance
```

```python
df_predictiveBSS = pd.read_pickle(r"../Results/predictive_bss_sound_separation_5by3_results.pkl")
df_predictiveBSS.tail(30)
```

```python
data = np.vstack(df_predictiveBSS["SNR"].values)

n = data.shape[0]
means = np.mean(data, axis=0)
sems = stats.sem(data, axis=0)

# 95% Confidence Interval Margin
# stats.t.ppf(0.975, df) gives the t-critical value
t_crit = stats.t.ppf(0.975, n-1)
ci_margins = t_crit * sems

for i in range(3):
    print(f"Sound {i+1}: {means[i]:.2f} +/- {ci_margins[i]:.2f} (95% CI)")
```
